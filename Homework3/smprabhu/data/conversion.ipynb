```python
# Collating data continent wise
import pandas as pd
import numpy as np
import csv
import json

df = pd.read_csv("bar.csv")
continents = df["Region"].unique()
years = [1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2019]
headers = ["Region","Year","Emissions Per Capita"]


with open("continents.csv", "a") as fp:
    writer = csv.writer(fp)
    writer.writerow(headers)
    for c in continents:
        for y in years:
            data = df[(df["Region"] == c) & (df["Year"] == y)]
            csv_row = [c, y, data["Emissions Per Capita"].mean() or 0.0]
            writer.writerow(csv_row)


```

```python
country_code = 0
with open("output.csv", "a") as fp:
    writer = csv.writer(fp)
    writer.writerow(headers)
    
    for a in countries:

        country_region = df.loc[df['Country Name'] == a]["Region"].values[0]
        for b in cols:
            b = str(b)
            data = df.loc[df['Country Name'] == a][str(b)]
            # print(data.values[0])
            
            csv_row = [a, country_code, country_region, country_region, b, data.values[0], 1]
            writer.writerow(csv_row)
            # print(type(data))
        country_code += 1
```
