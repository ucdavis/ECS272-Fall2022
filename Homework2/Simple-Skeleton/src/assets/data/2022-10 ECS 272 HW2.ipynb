# Process Data To Find Movies And Extract Primary Genre

```python
movies = titleData.loc[titleData['type']=='MOVIE'].reset_index(drop=True)
```

```python
def remove_extra_characters(string):
    return string.replace("[","").replace("]","").replace("'","")
```

```python
def get_primary(string):
    return string.split(', ')[0]
```

```python
movies['genres']=movies['genres'].apply(remove_extra_characters)
```

```python
movies['primary genre'] = movies['genres'].apply(get_primary)
```

```python
movies['production_countries']=movies['production_countries'].apply(remove_extra_characters)
```

```python
movies['primary country']=movies['production_countries'].apply(get_primary)
```

# Find count of secondary and tertiary genres for radar graph

```python
comedies
```

```python
secondaryTertiary=[]
for movie in comedies['title'].unique():
    movieGenres = comedies.loc[comedies['title']==movie,'genres'].values.item()
    genres = movieGenres.split(', ')
    if len(genres)==2:
        secondaryTertiary.append(genres[1])
    elif len(genres)>2:
        secondaryTertiary.append(genres[1])
        secondaryTertiary.append(genres[2])
```

```python
def countX(lst, x):
    return lst.count(x)
```

```python
uniqueGenres = np.unique(np.array(secondaryTertiary))
countSecondaryTertiary = []
for genre in uniqueGenres:
    countSecondaryTertiary.append({'axis':genre,'value':countX(secondaryTertiary,genre)})
```

```python
countOtherGenresOut = {"data":countSecondaryTertiary}
```

```python
countOtherGenresOut
```

```python
with open("countOtherGenresOut.json", "w") as final:
    json.dump(countOtherGenresOut, final)
```

# Redo analysis for radar plot to get director for each movie

```python
prolificDirectors = ['Ryan Polito','Jay Karas','Jay Chapman','Marcus Raboy','RaÃºl Campos','Jan Suter']
```

```python
directorGenreOverlap = []
for director in prolificDirectors:
    directorProfile = creditData.loc[creditData['name']==director]
    directorMovies=[]
    for movie in directorProfile['id'].unique():
        directorMovies.append(movies.loc[(movies['id']==movie)&(movies['primary genre']=='comedy')])
    directorMovies = pd.concat(directorMovies).reset_index(drop=True)
    otherGenres=[]
    for movie in directorMovies['title'].unique():
        movieGenres = comedies.loc[comedies['title']==movie,'genres'].values.item()
        genres = movieGenres.split(', ')
        for genre in genres:
            if genre!='comedy':
                otherGenres.append(genre)
    directorGenreOverlap.append({'director':director,'genre':otherGenres})
    
```
