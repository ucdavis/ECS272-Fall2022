```python
decade = {}

for i in range(1920, 2021, 10):
    decade[i] = {}

# print(decade)


for index, row in df3.iterrows():
    digit = row['year'] % 10
    year = row['year'] - digit
    innerMap = decade[year]
    
    if row['artists'] not in innerMap:
        innerMap[row['artists']] = 1
    else:
        innerMap[row['artists']] += 1
    
    decade[year] = innerMap
    
    
    
    


# for key, value in decade.items():
#     print(key, value)
#     print()
    
# for d in decade:
#     if d == 1930:
#         for y in decade[d]:
#     #         if decade[d][y] > 1:
#             print(d, y, decade[d][y])

# for d in decade:
#     print(max(decade[d]))
#     for a in decade[d]:
# #         print(d, a, decade[d][y])
#         print(max(d, a, decade[d][y]))
```

```python
listy = []
for index, row in df3.iterrows():
#     digit = row['year'] % 10
#     diff = 9 - digit
#     begin = row['year'] - digit
#     end = row['year'] + diff
    
    if (row['artists'] in artistYear):
        beg = artistYear[row['artists']]
        end = artistYear[row['artists']] + 9
        if (begin <= row['year'] <= end):
            r = row.to_dict()
            listy.append(r)
df4 = pd.DataFrame(listy)
df4
```

```python
filter = {"['Billie Holiday']": 10, "['Orchestra Studio 7']": 10, "['Miles Davis']": 10, "['The Beach Boys']": 10, "['Queen']": 10, "['Metallica']": 10, "['Nirvana']": 10,  "['Eminem']": 10, "['Taylor Swift']": 10, "['Lil Uzi Vert']": 10}
```

```python
listy = []
for index, row in df4.iterrows():
    if (row['artists'] in filter) and (filter[row['artists']]) > 0:
        filter[row['artists']] -= 1
        digit = row['year'] % 10
        row['year'] = row['year'] - digit
        r = row.to_dict()
        listy.append(r)
             
df5 = pd.DataFrame(listy)
df5
```
